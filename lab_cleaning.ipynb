```python
import pandas as pd

lab = pd.read_csv("ALL_DATA/lab.csv")
lab.head()
```

```python
lab = lab[['patientunitstayid', 'labresultoffset', 'labname', 'labresult']]
lab.head()
```

```python
lab_pivot = lab.pivot_table('labresult', ['patientunitstayid', 'labresultoffset'], 'labname')
```

```python
lab_pivot.head(20)
```

```python
lab_pivot = lab_pivot[["albumin", "BUN", "bicarbonate", "creatinine", "FiO2", "glucose", "HCO3", "Hct", "paO2", "pH", "potassium", "Respiratory Rate", "sodium", "Temperature", "total bilirubin", "WBC x 1000"]]
lab_pivot.head(20)
```

```python
lab_pivot = lab_pivot.query('labresultoffset > 0')
lab_pivot.head()
```

```python
lab_pivot
```

```python
lab_pivot = lab_pivot.reset_index(level=[0,1])
```

```python
lab_pivot = lab_pivot.drop(['labresultoffset'], axis=1)
```

```python
lab_pivot = lab_pivot.drop_duplicates(subset=['patientunitstayid'], keep='last')
```

```python
lab_pivot["fiO2/paO2 ratio"] = lab_pivot.FiO2/lab_pivot.paO2
```

```python
lab_pivot
```

```python
#save to csv file
lab_pivot.to_csv("lab1.csv")
```
